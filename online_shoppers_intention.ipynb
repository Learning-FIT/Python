# 課題（3）

```python
from sklearn.preprocessing import LabelEncoder

df_new = df[df['VisitorType'] == 'New_Visitor'].drop(['PageValues'], axis=1)

for category in ['Month']:
    le = LabelEncoder()
    le.fit(df_new[category])
    df_new[category] = le.transform(df_new[category])
    
df_new
```

```python
print('新規ユーザー')
plt.figure(figsize=(15, 15))
sns.heatmap(df_new.corr(), annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder

df_re = df[df['VisitorType'] == 'Returning_Visitor'].drop(['PageValues'], axis=1)

for category in ['Month']:
    le = LabelEncoder()
    le.fit(df_re[category])
    df_re[category] = le.transform(df_re[category])
    
print('再訪ユーザー')
plt.figure(figsize=(15, 15))
sns.heatmap(df_re.corr(), annot=True)
```

```python
df_new_we = df_new[df_new['Weekend'] == True].drop(['Weekend'], axis=1)
df_new_we
```

```python
print('新規ユーザー：週末')
plt.figure(figsize=(15, 15))
sns.heatmap(df_new_we.corr(), annot=True)
```

```python
df_new_wd = df_new[df_new['Weekend'] == False].drop(['Weekend'], axis=1)
print('新規ユーザー：平日')
plt.figure(figsize=(15, 15))
sns.heatmap(df_new_wd.corr(), annot=True)
```

```python
df_re_we = df_re[df_re['Weekend'] == True].drop(['Weekend'], axis=1)
print('再訪ユーザー：週末')
plt.figure(figsize=(15, 15))
sns.heatmap(df_re_we.corr(), annot=True)
```

```python
df_re_wd = df_re[df_re['Weekend'] == False].drop(['Weekend'], axis=1)
print('再訪ユーザー：平日')
plt.figure(figsize=(15, 15))
sns.heatmap(df_re_wd.corr(), annot=True)
```

```python
df_new_sd = df_new[df_new['SpecialDay'] > 0].drop(['SpecialDay'], axis=1)
df_new_sd
```

```python
print('新規ユーザー：特別日')
plt.figure(figsize=(15, 15))
sns.heatmap(df_new_sd.corr(), annot=True)
```
