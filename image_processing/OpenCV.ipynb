```python
import cv2

img = cv2.imread('dog.jpg')
height = img.shape[0]
width = img.shape[1]

if width < height:
    new_width = 100
    ratio = new_width / width
    new_height = int(height * ratio)
    left = 0
    right = 100
    upper = int((new_height - 100) / 2)
    lower = upper + 100
else:
    new_height = 100
    ratio = new_height / height
    new_width = int(width * ratio)
    left = int((new_width - 100) / 2)
    right = left + 100
    upper = 0
    lower = 100

img2 = cv2.resize(img, (new_width, new_height))
img3 = img2[upper:lower, left:right]
img3.shape
cv2.imwrite('dog_thumb.jpg', img3)
```

```python
import numpy as np

img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols, channel = img.shape

# 元画像
orig = np.float32([[0, 0], [cols, 0], [0, rows], [cols, rows]])
# 透視投影後の画像
pers = np.float32([[100, 0], [cols-100, 0], [0, rows], [cols, rows]])

M = cv2.getPerspectiveTransform(orig, pers)
img_pers = cv2.warpPerspective(img, M, (cols, rows))

plt.imshow(img_pers)
plt.show()
```

```python
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print('img:', img.dtype)

img_lap1 = cv2.Laplacian(img, ddepth=-1)
print('img_lap1:', img_lap1.dtype)
plt.imshow(img_lap1)
plt.show()

img_lap2 = cv2.Laplacian(img, ddepth=cv2.CV_64F)
print('img_lap2:', img_lap2.dtype)
plt.imshow(img_lap2)
plt.show()
```

```python
# 画像を左右反転した後で、NumPyを使ってグレースケール化してください。

img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_flip2 = cv2.flip(img, 1)
img_gray = 0.299*img_flip2[:,:,2] + 0.587*img_flip2[:,:,1] + 0.114*img_flip2[:,:,0]

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# 画像を135度で回転した後で、大津の手法で2値化してください。

img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

center = (img.shape[1] / 2, img.shape[0] / 2)
size = (img.shape[1], img.shape[0])

rotate = cv2.getRotationMatrix2D(center, 135.0, 1.0)
img_135 = cv2.warpAffine(img, rotate, size, flags=cv2.INTER_CUBIC)

img_gray = cv2.cvtColor(img_135, cv2.COLOR_RGB2GRAY)
ret, img_th = cv2.threshold(img_gray, 0, 255, cv2.THRESH_OTSU)

plt.imshow(img_th, cmap='gray')
plt.show()
```

```python
import os

ANACONDA3_DIR = 'D:/anaconda3'
data_dir = os.path.join(ANACONDA3_DIR, 'Lib', 'site-packages', 'cv2', 'data')

cascade_path = os.path.join(data_dir, 'haarcascade_frontalface_default.xml')
#cascade_path = os.path.join(data_dir, 'haarcascade_frontalface_alt.xml')
#cascade_path = os.path.join(data_dir, 'haarcascade_frontalface_alt2.xml')
#cascade_path = os.path.join(data_dir, 'haarcascade_frontalface_alt_tree.xml')
#cascade_path = os.path.join(data_dir, 'haarcascade_eye.xml')
#cascade_path = os.path.join(data_dir, 'haarcascade_upperbody.xml')

cascade = cv2.CascadeClassifier(cascade_path)

img = cv2.imread('person_2.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

facerect = cascade.detectMultiScale(img_gray)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for face in facerect:
    x, y = face[0:2]
    width, height = face[0:2] + face[2:4]
    cv2.rectangle(img, (x, y), (width, height), (255, 0, 0), thickness=2)
    
plt.imshow(img)
plt.show()
```
