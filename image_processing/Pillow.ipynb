```python
from PIL import Image

img = Image.open('dog.jpg')
```

```python
print('format:', img.format)
print('size;', img.size)
print('mode:', img.mode)
```

```python
import numpy as np

print(type(img))

img_array = np.array(img)

print(type(img_array))
print(img_array)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_array)
plt.show()
```

```python
img_gray = img.convert('L')
plt.imshow(np.array(img_gray), cmap='gray')
plt.show()
```

```python
img_180 = img.rotate(180)
plt.imshow(np.array(img_180))
plt.show()
```

```python
img_gray.save('dog_gray.jpg')
img_180.save('dog_180.jpg')
```

```python
img_gray.save('dog_gray.png')
img_180.save('dog_180.png')
```

```python
img_small = img.resize((int(img.size[0] * 0.5), int(img.size[1] * 0.5)))
plt.imshow(np.array(img_small))
plt.show()
```

```python
img_crop = img.crop((100, 100, 400, 600))
plt.imshow(np.array(img_crop))
plt.show()
```

# 課題（1）

```python
from PIL import Image

img = Image.open('dog.jpg')
width = img.size[0]
height = img.size[1]

if width < height:
    new_width = 100
    ratio = new_width / width
    new_height = int(height * ratio)
    left = 0
    right = 100
    upper = int((new_height - 100) / 2)
    lower = upper + 100
else:
    new_height = 100
    ratio = new_height / height
    new_width = int(width * ratio)
    left = int((new_width - 100) / 2)
    right = left + 100
    upper = 0
    lower = 100
    
img_small = img.resize((new_width, new_height))

img_crop = img_small.crop((left, upper, right, lower))

print(img_crop.size)

plt.imshow(np.array(img_crop))
plt.show()
```
